```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

base_dir = "../data"

splits = ["train", "test"]
classes = ["REAL", "FAKE"]

for split in splits:
    for cls in classes:
        path = os.path.join(base_dir, split, cls)
        count = len(os.listdir(path))
        print(f"{split}/{cls}: {count}장")

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle("CIFAKE Sample Images", fontsize=16)

for row, cls in enumerate(["REAL", "FAKE"]):
    path = os.path.join(base_dir, "train", cls)
    files = os.listdir(path)[:5]
    for col, fname in enumerate(files):
        img = Image.open(os.path.join(path, fname))
        axes[row][col].imshow(img)
        axes[row][col].set_title(cls)
        axes[row][col].axis("off")

plt.tight_layout()
plt.show()
```
