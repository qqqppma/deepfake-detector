```python
import tarfile
import os

tar_path = r"C:\Users\user\Desktop\deepfake-detector\data\vidprom\pika_videos_1.tar"
extract_path = r"C:\Users\user\Desktop\deepfake-detector\data\vidprom\pika_fake"

os.makedirs(extract_path, exist_ok=True)

with tarfile.open(tar_path, "r") as tar:
    members = tar.getmembers()
    print(f"총 영상 개수: {len(members)}")
    for i, member in enumerate(members):
        tar.extract(member, extract_path)
        if (i+1) % 100 == 0:
            print(f"{i+1}/{len(members)} 완료")

print("압축 해제 완료!")
```

```python
import os

fake_path = r"C:\Users\user\Desktop\deepfake-detector\data\vidprom\pika_fake"

mp4_files = []
for root, dirs, files in os.walk(fake_path):
    for f in files:
        if f.endswith(".mp4"):
            mp4_files.append(os.path.join(root, f))

print(f"총 fake 영상 개수: {len(mp4_files)}")
```

```python
import subprocess
import os

save_path = r"C:\Users\user\Desktop\deepfake-detector\data\vidprom\real_videos"
os.makedirs(save_path, exist_ok=True)

queries = [
    "daily vlog", "travel vlog", "cooking vlog",
    "city walk", "nature walk", "street vlog"
]

for query in queries:
    print(f"다운로드 중: {query}")
    subprocess.run([
        "yt-dlp",
        f"ytsearch20:{query}",
        "--output", os.path.join(save_path, "%(id)s.%(ext)s"),
        "--format", "mp4",
        "--max-filesize", "100m",
        "--no-playlist",
        "--ignore-errors",  # 에러 무시하고 계속 진행
    ])

print("다운로드 완료!")
```
