```python
import cv2
import json
import os
from pathlib import Path
from datetime import datetime

BASE_DIR = Path(r"C:\Users\user\Desktop\deepfake-detector")
VIDEO_DIR = BASE_DIR / "data" / "real_videos"
FRAME_DIR = BASE_DIR / "data" / "real_frames"
PROGRESS_PATH = BASE_DIR / "data" / "real_frames_progress.json"

FRAME_DIR.mkdir(parents=True, exist_ok=True)
(FRAME_DIR / "documentary").mkdir(exist_ok=True)
(FRAME_DIR / "vlog").mkdir(exist_ok=True)

FRAMES_PER_VIDEO = 1000
FRAME_INTERVAL = 5

print(f"프레임 저장 경로: {FRAME_DIR}")
```

```python
def load_progress():
    if PROGRESS_PATH.exists():
        with open(PROGRESS_PATH, "r") as f:
            return json.load(f)
    return {"done": []}

def save_progress(progress):
    with open(PROGRESS_PATH, "w") as f:
        json.dump(progress, f)

progress = load_progress()
print(f"이전 진행: {len(progress['done'])}개 완료")
```

```python
def extract_frames(video_path, save_dir, video_id, frames_per_video, frame_interval):
    cap = cv2.VideoCapture(str(video_path))
    if not cap.isOpened():
        return 0
    
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    saved = 0
    frame_idx = 0
    
    while saved < frames_per_video and frame_idx < total_frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
        ret, frame = cap.read()
        if not ret:
            break
        save_path = save_dir / f"{video_id}_frame{saved:04d}.jpg"
        cv2.imwrite(str(save_path), frame)
        saved += 1
        frame_idx += frame_interval
    
    cap.release()
    return saved
```

```python
all_videos = (
    [(p, "documentary") for p in sorted((VIDEO_DIR / "documentary").glob("*.mp4"))] +
    [(p, "vlog")        for p in sorted((VIDEO_DIR / "vlog").glob("*.mp4"))]
)

print(f"총 영상: {len(all_videos)}개")
total_frames = 0
start_time = datetime.now()

for i, (video_path, category) in enumerate(all_videos):
    vid_id = video_path.stem
    
    if vid_id in progress["done"]:
        print(f"[{i+1}/{len(all_videos)}] 스킵: {vid_id}")
        continue
    
    print(f"\n[{i+1}/{len(all_videos)}] 추출 중: {video_path.name}")
    
    saved = extract_frames(
        video_path,
        FRAME_DIR / category,
        vid_id,
        FRAMES_PER_VIDEO,
        FRAME_INTERVAL
    )
    
    total_frames += saved
    progress["done"].append(vid_id)
    save_progress(progress)
    
    elapsed = (datetime.now() - start_time).seconds // 60
    print(f"  ✓ {saved}프레임 저장 | 누적: {total_frames}장 | 경과: {elapsed}분")

print(f"\n완료! 총 {total_frames}장 추출")
```
